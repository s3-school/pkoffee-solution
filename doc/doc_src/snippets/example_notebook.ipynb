```python
from pathlib import Path

from pkoffee.data import load_csv

data = load_csv(Path("../example_data/coffee_productivity.csv"))
data.head()
```

```python
from pkoffee.productivity_analysis import fit_all_models
from pkoffee.visualization import format_model_rankings

models = fit_all_models(data)
print(format_model_rankings(models))  # noqa: T201
```

```python
from pkoffee.visualization import plot_models

plot_models(data, models)
```
